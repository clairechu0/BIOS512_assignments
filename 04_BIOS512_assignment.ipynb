```r
library('tidyverse')
```

```r
wind_data = read_csv(
    'https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-10-27/wind-turbine.csv', 
    col_types = cols())

wind_data |> head()
```

Each row in the table represents a single turbine.

Make a bar chart ğŸ“Š **sorted by frequency** of turbine count per province territory. (Hint: `fct_infreq` will sort your factor levels by frequency for subsequent plotting)

```r
wind_data |> 
    pull('province_territory') |>
    fct_infreq()
```

```r
options(repr.plot.width = 10, repr.plot.height = 10)

df_sorted = wind_data |>
    mutate(province_sorted = province_territory |> fct_infreq())

p = ggplot(df_sorted, aes(x = province_sorted)) +
    geom_bar() + 
    labs(x = 'Province Territory') +
    theme(axis.text.x = element_text(angle = 45))

p
```

Make a bar chart **sorted by frequency** of turbine count per manufacturer. **Group the low frequency manufacturers into an `Other` category**.

**Bonus:** Can you force the `Other` category to be the last bar? 

```r
options(repr.plot.width = 10, repr.plot.height = 10)

df_lumped = wind_data |>
    mutate(manufacturer_lumped = manufacturer |> fct_lump(n = 5))

df_sorted = df_lumped |>
    mutate(manufacturer_sorted = manufacturer_lumped |> fct_infreq())

p = ggplot(df_sorted, aes(x = manufacturer_sorted)) +
    geom_bar() + 
    labs(x = 'Manufacturer') +
    theme(axis.text.x = element_text(angle = 45))

p
```
